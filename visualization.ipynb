```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 
sns.set(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(15, 18))

# --------------------------
# 1
# --------------------------
plt.subplot(3, 2, 1)
gender_data = {
    'Male': {'N': 6641, 'Age': 51.23, 'Infection': 47.85},
    'Female': {'N': 7560, 'Age': 53.94, 'Infection': 48.77}
}

x = np.arange(2)
width = 0.35

bars1 = plt.bar(x - width/2, [gender_data['Male']['Infection'], gender_data['Female']['Infection']], 
                width, label='COVID Rate (%)', color='#3498db')
bars2 = plt.bar(x + width/2, [gender_data['Male']['Age'], gender_data['Female']['Age']], 
                width, label='Mean Age', color='#e74c3c')

plt.title('Gender Comparison')
plt.xticks(x, ['Male (N=6,641)', 'Female (N=7,560)'])
plt.ylabel('Percentage / Years')
plt.legend()

# 
for bar in bars1 + bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}',
             ha='center', va='bottom')

# --------------------------
# 2
# --------------------------
plt.subplot(3, 2, 2)
ethnicity_data = {
    'Ethnicity': ['White (N=10,788)', 'Black (N=2,154)', 'Asian (N=1,321)', 'Hispanic (N=1,245)'],
    'Infection Rate': [46.89, 50.46, 46.40, 53.25]
}

bars = plt.barh(ethnicity_data['Ethnicity'], ethnicity_data['Infection Rate'], 
                color=['#2ecc71', '#f39c12', '#9b59b6', '#1abc9c'])
plt.title('Infection Rate by Ethnicity')
plt.xlabel('Percentage (%)')

# 
for bar in bars:
    width = bar.get_width()
    plt.text(width + 0.5, bar.get_y() + bar.get_height()/2,
             f'{width:.2f}%',
             va='center')

# --------------------------
# 3
# --------------------------
plt.subplot(3, 2, (3,4))
region_data = {
    'Region': ['Northeast', 'Midwest', 'South', 'West'],
    'Sample Size': [3300, 2900, 4240, 3761],
    'Infection Rate': [47.42, 49.17, 48.55, 45.86]
}

# 
plt.subplot(3, 2, 3)
plt.pie(region_data['Sample Size'], labels=region_data['Region'], 
        autopct='%1.1f%%', colors=sns.color_palette("pastel"),
        startangle=90)
plt.title('Sample Distribution by Region')

# 
plt.subplot(3, 2, 4)
bars = plt.bar(region_data['Region'], region_data['Infection Rate'],
               color=sns.color_palette("husl", 4))
plt.title('Infection Rate by Region')
plt.ylabel('Percentage (%)')

# 
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2f}%',
             ha='center', va='bottom')

# --------------------------
# 4
# --------------------------
#  - 
plt.subplot(3, 2, 5)
age_data = pd.DataFrame({
    'Gender': ['Male']*1000 + ['Female']*1000,  # 
    'Age': np.concatenate([
        np.random.normal(51.23, 5, 1000),
        np.random.normal(53.94, 5, 1000)
    ])
})

sns.boxplot(x='Gender', y='Age', data=age_data, palette=["#3498db", "#e74c3c"])
plt.title('Age Distribution (Sampled 1000 each)')

# --------------------------
# 5
# --------------------------
plt.subplot(3, 2, 6)
plt.plot(region_data['Region'], region_data['Infection Rate'], 
         marker='o', color='#34495e', linewidth=2.5)
plt.title('Regional Infection Trend')
plt.ylabel('Percentage (%)')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# 
sns.set(style="whitegrid", font_scale=1.1)
plt.figure(figsize=(15, 10))

# --------------------------
# 
# --------------------------
fatigue_data = {
    'Gender': {
        'Male': 19.6,
        'Female': 23.1
    },
    'Ethnicity': {
        'Hispanic': 25.7,
        'Black': 22.0,
        'White': 20.3
    },
    'Region': {
        'South': 24.0,
        'West': 18.2
    }
}

# 1
plt.subplot(2, 3, 1)
sns.barplot(x=list(fatigue_data['Gender'].keys()), 
            y=list(fatigue_data['Gender'].values()),
            palette=["#3498db", "#e74c3c"])
plt.title('A) Fatigue by Gender')
plt.ylabel('Percentage (%)')
for i, v in enumerate(fatigue_data['Gender'].values()):
    plt.text(i, v+0.5, f"{v}%", ha='center')

# 2
plt.subplot(2, 3, 2)
sns.barplot(x=list(fatigue_data['Ethnicity'].keys()), 
            y=list(fatigue_data['Ethnicity'].values()),
            palette=["#2ecc71", "#f39c12", "#9b59b6"])
plt.title('B) Fatigue by Ethnicity')
plt.ylabel('Percentage (%)')
for i, v in enumerate(fatigue_data['Ethnicity'].values()):
    plt.text(i, v+0.5, f"{v}%", ha='center')

# 3
plt.subplot(2, 3, 3)
sns.barplot(x=list(fatigue_data['Region'].keys()), 
            y=list(fatigue_data['Region'].values()),
            palette=["#34495e", "#1abc9c"])
plt.title('C) Fatigue by Region')
plt.ylabel('Percentage (%)')
for i, v in enumerate(fatigue_data['Region'].values()):
    plt.text(i, v+0.5, f"{v}%", ha='center')

# --------------------------
# 
# --------------------------
longcovid_data = {
    'Gender': {
        'Male': 6.7,
        'Female': 8.4
    },
    'Ethnicity': {
        'Black': 9.2,
        'White': 7.1
    },
    'Region': {
        'South': 8.1,
        'West': 6.2
    }
}

# 4
plt.subplot(2, 3, 4)
sns.barplot(x=list(longcovid_data['Gender'].keys()), 
            y=list(longcovid_data['Gender'].values()),
            palette=["#3498db", "#e74c3c"])
plt.title('D) Long COVID by Gender\n(Among COVID-positive)')
plt.ylabel('Percentage (%)')
for i, v in enumerate(longcovid_data['Gender'].values()):
    plt.text(i, v+0.3, f"{v}%", ha='center')

# 5
plt.subplot(2, 3, 5)
sns.barplot(x=list(longcovid_data['Ethnicity'].keys()), 
            y=list(longcovid_data['Ethnicity'].values()),
            palette=["#f39c12", "#9b59b6"])
plt.title('E) Long COVID by Ethnicity\n(Among COVID-positive)')
plt.ylabel('Percentage (%)')
for i, v in enumerate(longcovid_data['Ethnicity'].values()):
    plt.text(i, v+0.3, f"{v}%", ha='center')

# 6
plt.subplot(2, 3, 6)
sns.barplot(x=list(longcovid_data['Region'].keys()), 
            y=list(longcovid_data['Region'].values()),
            palette=["#34495e", "#1abc9c"])
plt.title('F) Long COVID by Region\n(Among COVID-positive)')
plt.ylabel('Percentage (%)')
for i, v in enumerate(longcovid_data['Region'].values()):
    plt.text(i, v+0.3, f"{v}%", ha='center')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 
sns.set(style="whitegrid", font_scale=1.2)
plt.figure(figsize=(14, 6))

# 
bmi_data = {
    'Category': ['Underweight', 'Normal', 'Overweight', 'Obese'],
    'Fatigue (%)': [25.5, 20.3, 21.7, 24.6],
    'Color': ['#66c2a5', '#8da0cb', '#e78ac3', '#fc8d62']
}

smoking_data = {
    'Status': ['Never', 'Former', 'Current'],
    'Fatigue (%)': [20.1, 22.5, 27.2],
    'Color': ['#a6d854', '#66c2a5', '#e78ac3']
}

# --------------------------
# 1BMI
# --------------------------
plt.subplot(1, 2, 1)
bars1 = plt.bar(bmi_data['Category'], bmi_data['Fatigue (%)'], 
               color=bmi_data['Color'], alpha=0.8)
plt.title('Fatigue by BMI Category', pad=20)
plt.xlabel('BMI Category', labelpad=10)
plt.ylabel('Percentage Reporting Fatigue (%)', labelpad=10)
plt.ylim(0, 30)

# 
for bar in bars1:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}%', ha='center', va='bottom', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# --------------------------
# 2
# --------------------------
plt.subplot(1, 2, 2)
bars2 = plt.bar(smoking_data['Status'], smoking_data['Fatigue (%)'], 
               color=smoking_data['Color'], alpha=0.8)
plt.title('Fatigue by Smoking Status', pad=20)
plt.xlabel('Smoking Status', labelpad=10)
plt.ylabel('Percentage Reporting Fatigue (%)', labelpad=10)
plt.ylim(0, 30)

# 
for bar in bars2:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height}%', ha='center', va='bottom', fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# 
plt.tight_layout(pad=3.0)

# 
plt.suptitle('Fatigue Symptoms by Health Characteristics', y=1.05, fontsize=16, fontweight='bold')

plt.show()
```
